```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import linear_model
from sklearn import linear_model
import statsmodels.api as sm

from scipy.stats import jarque_bera
from scipy.stats import normaltest

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format
```

```python
weather = pd.read_csv('/home/buket/Jupyter/Data/szeged-weather/weatherHistory.csv')
weather.head(10)
```

```python
Y = weather["Temperature (C)"] - weather["Apparent Temperature (C)"]
X = weather[["Humidity","Wind Speed (km/h)"]]
```

```python
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
```

```python
print('Değişkenler: \n', lrm.coef_)
print('Sabit değer (bias): \n', lrm.intercept_)
```

```python
import statsmodels.api as sm

X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

```python
weather["humidity&windspeed"]=weather["Humidity"]*weather['Wind Speed (km/h)']
X = weather[["Humidity","Wind Speed (km/h)","humidity&windspeed"]]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```

```python
X = weather[["Humidity","Wind Speed (km/h)","Visibility (km)"]]
X = sm.add_constant(X)

results = sm.OLS(Y,X).fit()
results.summary()
```
